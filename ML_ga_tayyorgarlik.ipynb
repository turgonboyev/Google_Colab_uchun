```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn

df = pd.read_csv('https://github.com/ageron/handson-ml2/blob/master/datasets/housing/housing.csv?raw=true')
df.head()
```

```python
from sklearn.model_selection import train_test_split
train_set, test_set = train_test_split(df, test_size=0.2, random_state=42)
```

```python
housing = train_set.drop('median_house_value', axis=1)
housing_label = train_set['median_house_value'].copy()
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='median')
```

```python
housing_num = housing.drop('ocean_proximity', axis=1)
housing_num.head()
```

```python
x = imputer.fit_transform(housing_num)
x
```

```python
housing_tr = pd.DataFrame(x, columns=housing_num.columns, index=housing_num.index)
housing_tr.head()
```

```python
housing_cat = housing[['ocean_proximity']]
housing_cat.value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder
cat_encoder = OneHotEncoder()

housing_1hot = cat_encoder.fit_transform(housing_cat)
housing_1hot.toarray()
```
