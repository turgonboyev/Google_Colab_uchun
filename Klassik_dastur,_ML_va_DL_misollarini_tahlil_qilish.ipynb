```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt

data = make_regression(n_samples=40, n_features=1, noise=10, random_state=42)

x_data, y_data = data[0], data[1]

plt.scatter(x_data, y_data)
```

```python
def classic_prog(x): return x*70

y_pred = [classic_prog(x) for x in x_data]

plt.scatter(x_data, y_data, label="Haqiqiy data sets")
plt.plot(x_data, y_pred, 'r', label="Dastur")

plt.xlabel("X")
plt.ylabel("Y")

plt.title("x*70")
plt.show()
```

```python
def weight_prog(x, w): return w*x*70

w=1.39

y_pred = [weight_prog(x, w) for x in x_data]

plt.scatter(x_data, y_data, label="Haqiqiy data sets")
plt.plot(x_data, y_pred, 'r', label="Dastur")

plt.xlabel("X")
plt.ylabel("Y")

plt.title("w*x*70")
plt.show()
```
